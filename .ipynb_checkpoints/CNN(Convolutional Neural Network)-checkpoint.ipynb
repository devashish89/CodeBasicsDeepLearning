```python
import numpy as np
import pandas as pd
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
(X_train,y_train), (X_test,y_test)=datasets.cifar10.load_data()
print(len(X_train), len(X_test))
```

```python
X_train.shape
```

```python
plt.figure(figsize=(2,2))
plt.imshow(X_train[1])
```

```python
print(y_train[1])
```

```python
# Scaling

X_train_scaled= X_train/255
X_test_scaled=X_test/255
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

# ANN - Artificial Neural Network

```python
# ANN - Artificial Neural Network

ann = keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(3000, activation="relu"),
    keras.layers.Dense(1000,activation="relu"),
    keras.layers.Dense(10,activation='sigmoid')
])

ann.compile(optimizer="SGD",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
ann.fit(X_train_scaled,y_train.flatten(),epochs=50)
```

```python
ann.evaluate(X_test_scaled,y_test)
```

```python
y_pred = ann.predict(X_test_scaled)
y_pred
```

```python
classes[np.argmax(y_pred[0])], classes[y_test[0][0]]
```

```python
y_test
```

```python
y_test.reshape(-1,)
```

```python
y_pred_classes = [np.argmax(ele) for ele in y_pred]
y_pred_classes
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred_classes)
print(cm)
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

print(classification_report(y_test, y_pred_classes))
```

# CNN : Convolutional Neural Network

```python
# CNN
cnn = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
    
cnn.compile(optimizer="SGD",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
cnn.fit(X_train_scaled,y_train,epochs=50)
```

```python
cnn.evaluate(X_test_scaled,y_test)
```

```python
y_pred = cnn.predict(X_test_scaled)
y_pred[0]
```

```python
y_pred_classes = [np.argmax(ele) for ele in y_pred]
y_pred_classes
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred_classes)
print(cm)
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

print(classification_report(y_test, y_pred_classes))
```
